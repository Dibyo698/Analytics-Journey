1)What is a random variable in probability theory
> A random variable in probability theory is a function that assigns a numerical value to each possible outcome of a random experiment.

2) What are the types of random variables?
> Two types:- Discrete Random variable,
Continuous Random Variables

3) What is the difference between discrete and continuous distributions?
> Discrete Distribution:- Describe the probability of outcomes for a random variable that can take on only specific, separate values. Represented by a probability mass function (PMF). e.g:-Number of cars in a parking lot (can only be 0, 1, 2, ...).

> Continuous Distribution:-Describe the probability of outcomes for a random variable that can take on any value within a given interval or range.Represented by a probability density function (PDF). e.g:-Amount of rain in a month (can be any real number, e.g., 2.56 inches)

4)What are probability distribution functions (PDF)?
>A probability distribution function (PDF), more precisely called a probability density function, is a function that describes the likelihood or density that a continuous random variable will take a value within a particular range.

5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
>The PDF gives the relative likelihood of the variable near a given value, but not the probability at a single point (which is zero for continuous variables), whereas The CDF gives the total probability accumulated up to and including a value, always between 0 and 1.

>The PDF is used to find probabilities over intervals (via integration), while the CDF directly gives probabilities up to a point.

>The CDF is always non-decreasing and smooth for continuous variables, while the PDF can vary in shape

6)What is a discrete uniform distribution?
>A discrete uniform distribution is a probability distribution where a finite number of distinct outcomes are equally likely to occur. In this distribution, each possible value has the same probability.

7)What are the key properties of a Bernoulli distribution?
>A Bernoulli distribution is a simple way to describe situations where there are only two possible outcomes: something either happens or it doesn't.

>Each trial is independent; the outcome of one trial does not affect another

8)What is the binomial distribution, and how is it used in probability?
>A binomial distribution is a way to figure out how likely it is to get a certain number of successes (like heads in a coin toss) when we repeat the same experiment a set number of times, and each time we only have two possible results: success or failure.

To find the chance that a medicine that works 80% of the time is effective for exactly 7 out of 10 patients, we use the binomial probability formula.

9)What is the Poisson distribution and where is it applied?
>The Poisson distribution is widely used to predict or explain the probability of a specific number of rare or random events occurring within a fixed period or area, as long as those events happen independently and at a constant average rate.
e.g:- Number of customer calls received by a call center in an hour(Lamda=10, k=7),
Number of cars passing through a toll booth in a minute

10)What is a continuous uniform distribution?
>A continuous uniform distribution is a probability distribution where every value within a certain interval is equally likely to occur. It is sometimes called a rectangular distribution because its probability density function (PDF) forms a rectangle when graphed.

11)What are the characteristics of a normal distribution?
* The distribution is perfectly symmetric around its center.
* There is only one peak (highest point) in the distribution
* The mean, median, and mode are all equal and located at the center of the distribution
* The shape and location of the curve are determined by the mean (center) and the standard deviation (spread)
* Total Area Under Curve = 1( total probability represented by the curve is 1)
* The normal distribution is for continuous data, not discrete values

12)What is the standard normal distribution, and why is it important?
>The standard normal distribution is a special case of the normal (bell curve) distribution where the mean is 0 and the standard deviation is 1. This distribution is also called the z-distribution

* Any normal distribution (with any mean and standard deviation) can be converted into the standard normal distribution using a process called standardization. This allows you to compare values from different normal distributions on the same scale

* In the standard normal distribution, each value is measured in terms of standard deviations from the mean. These are called z-scores. A z-score tells you how far a value is from the average, making it easy to interpret and compare results

* Many statistical methods and hypothesis tests rely on the standard normal distribution as a foundation, especially when applying the Central Limit Theorem

13)What is the Central Limit Theorem (CLT), and why is it critical in statistics?
>The Central Limit Theorem says that if we take a lot of random samples from any population—no matter what the population looks like—and calculate the average (mean) of each sample, those averages will form a bell-shaped curve (normal distribution) as long as our samples are big enough

* The CLT allows statisticians to use the normal distribution to make inferences about population means, even when the underlying population is not normally distributed, as long as the sample size is large enough (commonly, n ≥ 30)
* sample means are approximately normally distributed, we can calculate probabilities and perform hypothesis tests

14)How does the Central Limit Theorem relate to the normal distribution?
>The CLT states that if we take large enough random samples from any population (no matter what shape the original population's distribution is), the distribution of the sample means will become approximately normal (bell-shaped) as the sample size increases

15) What is the application of Z statistics in hypothesis testing?
>Z statistics (or Z-scores) are widely used in hypothesis testing to determine whether a sample result is significantly different from a population parameter or from another sample.

>The calculated Z-score is compared to a critical value from the standard normal distribution (for example, ±1.96 for a 5% significance level in a two-tailed test).

If the absolute value of the Z-score is greater than the critical value, we reject the null hypothesis (meaning the sample provides enough evidence to say there is a significant difference).

If not, we fail to reject the null hypothesis (meaning there isn’t enough evidence to say there is a significant difference).

16) How do you calculate a Z-score, and what does it represent?
>A Z-score (also called a standard score) tells us how many standard deviations a specific value is from the mean of a dataset.

The formula is:    z=(x−μ)/σ

17)What are point estimates and interval estimates in statistics?
>Point Estimate:-
A point estimate is a single number that we use as our best guess for something about a whole group (population), based on just a small part of it (sample)

>Interval Estimate:-
An interval estimate gives us a range of numbers that we think the true answer falls within, instead of just one guess

18)What is the significance of confidence intervals in statistical analysis?
>Confidence intervals are crucial in statistical analysis because they provide a quantitative measure of the uncertainty or precision associated with a sample estimate, such as a mean or proportion.
>> Rather than reporting a single value, a confidence interval gives a range of values—bounded by a lower and upper limit—within which the true population parameter is likely to fall.

19)What is the relationship between a Z-score and a confidence interval?
>A Z-score and a confidence interval are closely related in statistics. The Z-score is a number that tells us how many standard deviations a value is from the mean in a standard normal distribution. When we create a confidence interval, we use a specific Z-score (also called the "critical value") that matches our desired confidence level.

20) How are Z-scores used to compare different distributions?
> Z-scores are used to standardize values from different distributions, making it possible to compare them even if the distributions have different means and standard deviations. A Z-score tells us how many standard deviations a value is from the mean of its own distribution.

>Standardization: By converting raw scores into Z-scores, each value is expressed on the same scale
>>Relative Standing: We can directly compare how "unusual" or "above average" a value is within its own group, and then compare that to values from other groups or distributions

21)What are the assumptions for applying the Central Limit Theorem?
>assumptions for using the Central Limit Theorem (CLT):
* Random Sampling: We need to pick our data randomly, so every item in the group has a fair chance of being chosen
* Independence: Each data point we pick shouldn’t affect the others. For example, picking one card from a deck shouldn’t change the chances for the next card
* Sample Size: our sample should be big enough. Usually, if we have 30 or more data points, that’s good for most situations
* 10% Rule (if not putting things back): If we’re picking without putting things back (like drawing marbles from a bag), don’t take more than 10% of the total group. This keeps your picks independent
* Finite Variance: The group we’re picking from shouldn’t have extreme outliers or endless spread.

22) What is the concept of expected value in a probability distribution?
>The expected value in a probability distribution is the long-term average or mean we would expect to get if we repeated a random experiment many times. It is calculated by multiplying each possible outcome by its probability, then adding up all those results. This gives us a single number that summarizes what we can expect “on average” from the random process.

23)How does a probability distribution relate to the expected outcome of a random variable?
>A probability distribution shows all the possible outcomes of a random variable and how likely each one is to happen. The expected outcome—also called the expected value or mean—is a number that tells us what we would get on average if we repeated the random process many times

The expected value is calculated by:

* Taking each possible outcome,
* Multiplying it by its probability,
* Adding up all those results.

The expected value is basically the "center" or average of the probability distribution.

3)Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return p if x == 1 else (1 - p)

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.6))  # Should return 0.6
print("P(X=0):", bernoulli_pdf(0, 0.6))  # Should return 0.4
```

9)Write a Python function to calculate Z-scores from a dataset and plot them.


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):

    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero; Z-scores cannot be computed.")

    z_scores = (data - mean) / std_dev
    return z_scores

data = np.random.normal(50, 10, 100)

z_scores = calculate_z_scores(data)

plt.figure(figsize=(8, 4))
plt.plot(z_scores, marker='o', linestyle='', color='blue')
plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.title("Z-scores of Dataset")
plt.grid(True)
plt.legend()
plt.show()
```

12)Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():

    x = np.linspace(-4, 4, 1000)

    pdf = norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, color='blue', linewidth=2, label='Standard Normal PDF')
    plt.fill_between(x, pdf, color='lightblue', alpha=0.5)

    for z in [-3, -2, -1, 0, 1, 2, 3]:
        plt.axvline(z, color='gray', linestyle='--', linewidth=1)
        plt.text(z, norm.pdf(z) + 0.01, f'z={z}', ha='center', fontsize=9)

    plt.title("Standard Normal Distribution (μ = 0, σ = 1)")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
plot_standard_normal_distribution()
```

15)Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

population_mean = 100
population_std = 15

sample = [110, 112, 108, 115, 109, 111, 113, 107, 114, 116]
sample_size = len(sample)
sample_mean = np.mean(sample)

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

reject_null = abs(z_stat) > z_critical

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"Critical Z-value: ±{z_critical:.4f}")
print("Result: ", "Reject Null Hypothesis (Significant Difference)" if reject_null else "Fail to Reject Null Hypothesis (No Significant Difference)")
```

16)Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

data = [52, 55, 60, 57, 58, 59, 61, 63, 62, 64]
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

confidence = 0.95
alpha = 1 - confidence

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

margin_of_error = t_critical * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Confidence Level: {confidence*100:.0f}%")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

24)Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_skewness_kurtosis(data):

    skew_value = skew(data)
    kurtosis_value = kurtosis(data)
    return skew_value, kurtosis_value

np.random.seed(1)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

skewness, excess_kurtosis = analyze_skewness_kurtosis(normal_data)

print(f"Skewness: {skewness:.4f}")
print(f"Excess Kurtosis: {excess_kurtosis:.4f}")
```
